```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use(r"C:\Users\B30724\Documents\Python\presentation.mplstyle")
```

```python
## Parameters
x = np.linspace(-1,1,101)
y = np.linspace(-1,1,101)
cutoff = 0.01
charge = -1

X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
R[R<cutoff] = cutoff
# Electric potential (cylindrical symmetry)
V = charge * np.log(R)

# Electric field E = -grad V
# Sub-sample x, y for quiver plot
x0 = x[::5]
y0 = y[::5]
X0, Y0 = np.meshgrid(x0, y0)
R0 = np.sqrt(X0**2 + Y0**2)
R0[R0<cutoff] = cutoff
theta = np.arctan2(Y0, X0)
Ex = -charge / R0 * np.cos(theta)
Ey = -charge / R0 * np.sin(theta)

fig, ax = plt.subplots(figsize=(5,4))
img = ax.pcolormesh(X, Y, V, cmap="viridis", vmax=3)  # vmin=-3 vmax=3
ax.quiver(X0, Y0, Ex, Ey, pivot='middle', headwidth=6)
ax.set(xlabel='X', ylabel='Y')
ax.set_aspect('equal')
fig.colorbar(img, ax=ax)
fig.tight_layout()
# fig.savefig("gradient_sink.png")
fig.savefig("gradient_source.png")
```
