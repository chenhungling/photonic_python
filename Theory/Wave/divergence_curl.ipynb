```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use(r"C:\Users\B30724\Documents\Python\presentation.mplstyle")
```

```python
def diverging_field(X, Y):
    u = X ** 2
    v = Y ** 2
    return u, v

x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)
u, v = diverging_field(X, Y)
# Color represents the divergence
color = 2*X + 2*Y

fig, ax = plt.subplots(figsize =(5,4))
img = ax.quiver(X, Y, u, v, color, clim=[-4, 4], cmap='vanimo', pivot='middle', headwidth=6)
ax.set(xlabel='X', ylabel='Y')
ax.set_aspect('equal')
fig.colorbar(img, ax=ax)
fig.tight_layout()
fig.savefig("field_diverging.png")
```

```python
def rotating_field(X, Y):
    u = np.sin(X) * np.cos(Y)
    v = - np.cos(X) * np.sin(Y)
    return u, v

x = np.linspace(-np.pi, np.pi, 20)
y = np.linspace(-np.pi, np.pi, 20)
X, Y = np.meshgrid(x, y)
u, v = rotating_field(X, Y)
# Color represents the curl
color = 2*np.sin(X)*np.sin(Y)

fig, ax = plt.subplots(figsize =(5,4))
img = ax.quiver(X, Y, u, v, color, clim=[-2, 2], cmap='vanimo', pivot='middle', headwidth=6)
ax.set(xlabel='X', ylabel='Y')
ax.set_aspect('equal')
fig.tight_layout()
fig.colorbar(img, ax=ax)
fig.savefig("field_rotating.png")
```
